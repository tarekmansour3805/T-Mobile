```python
capx=pd.read_csv('CAPX.csv')
```

```python
bom=pd.read_csv('bom.csv', low_memory=False)
```

```python
bom.dropna(subset=["Site Ref"], inplace=True)
```

```python
bom=bom[bom['Reference'].str.contains ("W4")]
```

```python
bom.drop(['Market', 'Project', 'Market<br/>Confirmed', 'Site', 'Unnamed: 15', 'Description', 'Shopping Cart', 'Cart Class', 'Project<br/>Unallocated Qty', 'Project Site<br/>Allocated Qty'], axis=1, inplace=True)
```

```python
bom=bom.groupby('SAP')['Plan To<br/>Deploy Qty','Site<br/>Allocated Qty', 'Allocation<br/>Shortage Qty'].sum()
```

```python
capx=capx[capx['Project'].str.contains ("CAPX")]
capx=capx[capx['Market'].str.contains ("E930 Dallas NDC")]
```

```python
capx.drop(['Manufacturer', 'Site Allocated Qty', 'Un-deployable Qty', 'Picked Qty', 'Total Planned Qty', 'Unnamed: 12', 'Total Planned Qty.1'], axis=1, inplace=True)
```
