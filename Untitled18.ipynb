```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pyodbc 
db = pyodbc.connect('Driver={SQL Server};'
                      'Server=PGI-M2M;'
                      'Database=M2MDATA04;'
                      'Trusted_Connection=yes;')
```

```python
query = 'select fpartno, fcpartrev, ffromlot, fctime_ts from dbo.intran'
```

```python
p = pd.read_sql_query(query , db)
```

```python
p['fpartno'] = p['fpartno'].str.replace(' ', '')
p['ffromlot'] = p['ffromlot'].str.replace(' ', '')
p['fctime_ts'] = p['fctime_ts'].replace(' ', '')
p['fcpartrev'] = p['fcpartrev'].str.replace(' ', '')

p.replace('', np.nan, inplace=True)
p.applymap(lambda x: x.rstrip() if type(x)==str else x)
```

```python
p = p[pd.notnull(p['ffromlot'])]
```

```python
p = p.drop_duplicates('ffromlot',keep = 'first')
```
