```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pyodbc 
db = pyodbc.connect('Driver={SQL Server};'
                      'Server=PGI-M2M;'
                      'Database=M2MDATA04;'
                      'Trusted_Connection=yes;')
```

```python
query = 'select ffromloc, FFROMBIN, fqty from dbo.intran'
```

```python
b = pd.read_sql_query(query , db)
```

```python
b['ffromloc'] = b['ffromloc'].str.replace(' ', '')
b['FFROMBIN'] = b['FFROMBIN'].str.replace(' ', '')
b['fqty'] = b['fqty'].replace(' ', '')
```

```python
b = b[(b['ffromloc']=='SR')]
b = b[b['fqty']>=1]
```

```python
b= b.drop_duplicates('FFROMBIN',keep = 'first')
b.replace('', np.nan, inplace=True)
b = b[pd.notnull(b['FFROMBIN'])]
b
```

```python
b2= ['#S'] + b['FFROMBIN'] + ['*1']
```

```python
#df = b.concat(b2, how='inner')
df = pd.concat([b, b2], axis=1, join='inner')
df
```

```python
df.to_csv('Bar Codes.csv', index = False)
```
