```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import pyodbc 
db = pyodbc.connect('Driver={SQL Server};'
                      'Server=PGI-M2M;'
                      'Database=M2MDATA04;'
                      'Trusted_Connection=yes;')
```

```python
df = pd.read_csv("Part Desc 2.csv")
query = 'select fpartno, ffromlot, FCTIME_TS, FQTY, ffromloc, FFROMBIN from dbo.intran'
```

```python
q = pd.read_sql_query(query , db)
```

```python
q['fpartno'] = q['fpartno'].str.replace(' ', '')
q['ffromlot'] = q['ffromlot'].str.replace(' ', '')
q['FCTIME_TS'] = q['FCTIME_TS'].replace(' ', '')
q['FQTY'] = q['FQTY'].replace(' ', '')
q['ffromloc'] = q['ffromloc'].str.replace(' ', '')


```

```python
q['fpartno'] = q['fpartno'].str.strip()
```

```python
q.replace('', np.nan, inplace=True)
```

```python
parts = df["fpartno"].tolist()
```

```python
q = q[q['fpartno'].isin(parts)]
```
